```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from tqdm import tqdm

# Paths to your data
train_csv = 'mitbih_train/mitbih_train.csv'
test_csv = 'mitbih_test/mitbih_test.csv'

# Read CSVs
train_df = pd.read_csv(train_csv, header=None)
test_df = pd.read_csv(test_csv, header=None)

# Create folders
def make_dirs(base_dir, labels):
    for label in labels:
        os.makedirs(os.path.join(base_dir, str(label)), exist_ok=True)

train_img_dir = 'train_images_ecg'
test_img_dir = 'test_images_ecg'
make_dirs(train_img_dir, range(5))
make_dirs(test_img_dir, range(5))

# Function to save waveform image
def save_ecg_plot(ecg_data, label, img_path):
    plt.figure(figsize=(2, 2), dpi=64)  # Makes 128x128 image
    plt.plot(ecg_data, color='black', linewidth=1)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(img_path, bbox_inches='tight', pad_inches=0)
    plt.close()

# Generate ECG plot images
def generate_ecg_images(df, output_dir, prefix='train'):
    for idx, row in tqdm(df.iterrows(), total=len(df), desc=f"Generating {prefix} ECG images"):
        label = int(row.iloc[-1])
        ecg_data = row.iloc[:-1]
        filename = f"{prefix}_{idx}.png"
        filepath = os.path.join(output_dir, str(label), filename)
        save_ecg_plot(ecg_data, label, filepath)

# Run generation
generate_ecg_images(train_df, train_img_dir, 'train')
generate_ecg_images(test_df, test_img_dir, 'test')

print(" ECG waveform image generation complete.")
```
